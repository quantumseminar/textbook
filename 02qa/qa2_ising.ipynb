## イジングモデル
物理学でイジングモデルは、磁石などの性質を表現するモデルで、上向と下向きのスピンと呼ばれる向きを持つ状態で表現される。

## イジングモデルとスピン（量子ビット）
今回計算で利用するスピン（量子ビット）は

上向きが +1、下向きが -1

の値を取る。このスピンをたくさん並べて計算を行う。

## スピンと量子ビット
スピンを量子ビットと呼ぶことができる。ビットとは、0か1の値を持つ計算の基本だが、量子ビットはちょっと違う性質を持っている。今回は+1もしくは-1の値を持つことができるが、+1と-1の状態を同時に持つ量子重ね合わせを実現できるものを使う。

今回量子ビットはたくさんあって、それぞれ単体の量子ビットにバイアスの値をかけると+1か-1になりやすくなるように設定できる。また、量子ビット同士の間に相互作用の値をかけると両方の量子ビットが同じ値になりやすいか違う値になりやすいかを決めることができる。

## ルール
いったん重ね合わせを忘れて、量子ビットには２つの値を設定できる。

・バイアス（局所磁場）
・ウェイト（相互作用）

![../img/qa2_02.jpg](https://github.com/quantumseminar/textbook/blob/main/img/qa2_02.jpg?raw=1)

バイアスは1量子ビットが+1（上向き）になりやすいか-1（下向き）になりやすいかを決めるために設定する値。

相互作用は2量子ビットが同じ向き (+1,+1) もしくは (-1,-1) になりやすいか、違う向き（+1, -1）もしくは (-1, +1)になりやすいかを決める値。

また、全体のルールとして、値にマイナスをつけた方が小さい方になりやすいように計算がされる。（イジングモデルはちょっとややっこしい。。。）

## 例題１：バイアス
量子ビット$q_0$がある。この量子ビットに+1の値をかけると

$$1*1=1$$

と

$$1*(-1)=-1$$

の両方の答えが考えられるが、量子ビットは値が小さい方になりやすいように計算されるので、-1が答えとして出る。量子ビットに1を設定すると-1が出やすい。

（セットアップによって、-1が出やすい場合と+1が出やすい場合があるので、セットアップによるが、この場合はそのまま計算する）

## 例題2:バイアス
量子ビット$q_0$がある。この量子ビットに-5.5の値をかけると

$$-5.5*1=-5.5$$

と

$$-5.5*(-1)=5.5$$

の両方の答えが考えられるが、量子ビットは値が小さい方になりやすいように計算されるので、+1が答えとして出る。量子ビットに-5.5を設定すると+1が出やすい。

## 例題3:相互作用
２つの量子ビット$q_0$と$q_1$がある。この量子ビットの間に+1の値をかけると

両方とも+1の場合（つまり同じ値の場合）、

$$1 * 1 * 1 = 1$$

両方とも-1の場合（つまり同じ値の場合）、

$$1 * (-1) * (-1) = 1$$

両方とも１になる。また、量子ビットの値がそれぞれ異なる場合、

$q_0 = +1$で、$q_1 = -1$の場合（つまり違う値の場合）、

$$1 * 1 * (-1) = -1$$

$q_0 = -1$で、$q_1 = +1$の場合（つまり違う値の場合）、

$$1 * (-1) * 1 = -1$$

なので、+1を相互作用としてかけると、お互いの量子ビットは異なる値になりやすくなる。


## 例題4:バイアスと相互作用の両方を使う場合
上記のバイアスと相互作用を同時に利用することもできる。$q_0$に+1のバイアスをかけ、$q_0$と$q_1$の相互作用に1を作用させた場合、

$q_0$にバイアス+1がかかっているので、$q_0$は-1になりやすい
$q_0$と$q_1$の相互作用には+1がかかっているので、$q_0$と$q_1$は違う値になりやすい。

ここで、$q_1$には値はかかっていないので、先に$q_0$を考えると、バイアスから$q_0$は-1に。そして、相互作用から違う値になるので、$q_1$は+1に。

なので、順番には考えると、$q_0=-1,q_1=1$に。

## 式の形に
上記を式の形にもできる。その際には最初に全部の式を書き出す。

$$
H = h_0 * q_0 + h_1 * q_1 + J_{01} * q_0 * q_1
$$

そして、$h_0 = 1$と$h_1=0$を代入して、$J_{01}=1$を代入します。すると、

$$
H = 1 * q_0 + 0 * q_1 + 1 * q_0 * q_1 = q_0 + q_0 * q_1
$$

となりました。この$q_0 + q_0 * q_1$が最小になるような$q_0$と$q_1$の組合せを考えます。
