## データビジュアライゼーション
量子コンピュータや機械学習では基本的にデータを利用します。データは理解しやすいように可視化（かしか）することがあります。ここでは、データを見やすくする方法を学びます。

ツールはmatplotlibを利用します。その中のpyplotというツールを使います。

```python
import matplotlib.pyplot as plt
```

## 折れ線チャート　Line Charts

今回は [1, 2, 3, 4] というリストをチャートにしてみます。

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

上のリストは縦のsome numbersに入りました。横は勝手に0,1,2,3が割り当てられます。
今度は横も縦も決めてみます。

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

今回は最初のリスト[1,2,3,4]が横軸xに対応し、次のリスト[1,4,9,16]が縦軸yに対応します。

## 点グラフ
線の代わりに点を打ってみます。ほとんど同じですが、デザインを変更します。

plotの中に三つ目の引数を設定することで、デザインを変更できます。

デフォルトでは'b-'となっているので、青線になります。今度は赤丸の'ro'にしてみます。

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis((0, 6, 0, 20))
plt.show()
```

点になりました。複数のラインやプロットをできます。

```python
# red dashes, blue squares and green triangles
plt.plot([0,1,2,3], [0,1,2,3], 'r--', [0,1,2,3], [0,1,4,9], 'bs', [0,1,2,3], [0,1,8,27], 'g^')
plt.show()
```

## 棒グラフ Bar Charts
次は棒グラフです。それぞれのxに対するyが棒になっています。

```python
plt.bar([0,1,2], [3,4,5])
plt.show()
```
